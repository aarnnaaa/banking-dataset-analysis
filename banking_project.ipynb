```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('default')
sns.set_palette("husl")

def load_data(file_path):
    try:
        df = pd.read_csv(file_path)
        print(f"dataset loaded")
        print(f"shape: {df.shape}")
        return df
    except Exception as e:
        print(f"error loading data: {e}")
        return None

def basic_info(df):
    print("basic dataset information")

    print(f"dataset shape: {df.shape}")
    print(f"missing values:\n{df.isnull().sum()}")
    print(f"\ndata types:\n{df.dtypes}")
    print(f"\nfirst 5 rows:")
    print(df.head())

def analyze_age_distribution(df):
    print("age analysis")

    print(f"age statistics:")
    print(f"mean: {df['age'].mean():.2f}")
    print(f"median: {df['age'].median():.2f}")
    print(f"min: {df['age'].min()}")
    print(f"max: {df['age'].max()}")

    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.hist(df['age'], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title('age Distribution')
    plt.xlabel('age')
    plt.ylabel('frequency')

    plt.subplot(1, 2, 2)
    plt.boxplot(df['age'])
    plt.title('age Box Plot')
    plt.ylabel('age')

    plt.tight_layout()
    plt.show()

def analyze_categorical_variables(df):
    categorical_cols = ['job', 'marital', 'education', 'default', 'housing',
                       'loan', 'contact', 'month', 'poutcome', 'y']

    print("categorical var analysis")


    for col in categorical_cols:
        if col in df.columns:
            print(f"\n{col.upper()} distribution:")
            value_counts = df[col].value_counts()
            percentages = df[col].value_counts(normalize=True) * 100

            for value, count in value_counts.items():
                print(f"  {value}: {count} ({percentages[value]:.1f}%)")
            plt.figure(figsize=(10, 6))
            if len(value_counts) <= 10:
                plt.bar(value_counts.index, value_counts.values, alpha=0.7)
                plt.xticks(rotation=45)
            else:
                plt.pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%')
            plt.title(f'{col.title()} Distribution')
            plt.tight_layout()
            plt.show()

def analyze_numerical_variables(df):
    numerical_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

    print("numerical var analysis")


    for col in numerical_cols:
        if col in df.columns:
            print(f"\n{col.upper()} stats:")
            print(f"  mean: {df[col].mean():.2f}")
            print(f"  median: {df[col].median():.2f}")
            print(f"  std: {df[col].std():.2f}")
            print(f"  min: {df[col].min():.2f}")
            print(f"  max: {df[col].max():.2f}")
    fig, axes = plt.subplots(3, 3, figsize=(15, 12))
    axes = axes.flatten()

    for i, col in enumerate(numerical_cols):
        if col in df.columns and i < 9:
            axes[i].hist(df[col], bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
            axes[i].set_title(f'{col.title()} distribution')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('frequency')
    for i in range(len(numerical_cols), 9):
        fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()

def analyze_target_variable(df):
    print("target var analysis")

    if 'y' in df.columns:
        target_counts = df['y'].value_counts()
        target_percentages = df['y'].value_counts(normalize=True) * 100

        print("term deposit subscription:")
        for value, count in target_counts.items():
            print(f"  {value}: {count} ({target_percentages[value]:.1f}%)")

        plt.figure(figsize=(10, 5))

        plt.subplot(1, 2, 1)
        plt.bar(target_counts.index, target_counts.values, alpha=0.7, color=['coral', 'lightblue'])
        plt.title('term deposit subscription distribution')
        plt.ylabel('Count')

        plt.subplot(1, 2, 2)
        plt.pie(target_counts.values, labels=target_counts.index, autopct='%1.1f%%',
                colors=['coral', 'lightblue'])
        plt.title('term deposit subscription %')

        plt.tight_layout()
        plt.show()

def analyze_correlations(df):
    print("correlation analsysi")

    df_encoded = df.copy()

    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        df_encoded[col] = pd.Categorical(df_encoded[col]).codes
    corr_matrix = df_encoded.corr()

    if 'y' in df_encoded.columns:
        target_corr = corr_matrix['y'].sort_values(key=abs, ascending=False)
        print("correlations with term deposit subscription (y):")
        for var, corr in target_corr.items():
            if var != 'y':
                print(f"  {var}: {corr:.3f}")
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
                square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
    plt.title('correlation matrix')
    plt.tight_layout()
    plt.show()

def campaign_analysis(df):
    print("campaign analysis")

    if 'campaign' in df.columns:
        print(f"campaign contacts statistics:")
        print(f"  average contacts per client: {df['campaign'].mean():.2f}")
        print(f"  max contacts: {df['campaign'].max()}")
        print(f"  clients with 1 contact: {(df['campaign'] == 1).sum()}")
        print(f"  clients with >5 contacts: {(df['campaign'] > 5).sum()}")

    if 'duration' in df.columns:
        print(f"\ncall duration statistics:")
        print(f"  average duration: {df['duration'].mean():.2f} seconds")
        print(f"  average duration: {df['duration'].mean()/60:.2f} minutes")

    if 'pdays' in df.columns:
        previously_contacted = df[df['pdays'] != -1]
        print(f"\nprevious campaign analysis:")
        print(f"  previously contacted clients: {len(previously_contacted)}")
        print(f"  never contacted before: {(df['pdays'] == -1).sum()}")

def success_factors_analysis(df):
    print("success factors analysis")

    if 'y' in df.columns:
        categorical_vars = ['job', 'marital', 'education', 'contact', 'month', 'poutcome']

        for var in categorical_vars:
            if var in df.columns:
                success_rate = df.groupby(var)['y'].apply(lambda x: (x == 'yes').mean() * 100)
                print(f"\nsuccess rate by {var}:")
                for category, rate in success_rate.sort_values(ascending=False).items():
                    print(f"  {category}: {rate:.1f}%")

def upload_file_colab():
    try:
        from google.colab import files
        print("please upload your Banking_data.csv file:")
        uploaded = files.upload()
        return list(uploaded.keys())[0]
    except ImportError:
        return None

def main():
    print("banking dataset analysis")

    try:
        import google.colab
        in_colab = True
        print("running in google colab")
    except ImportError:
        in_colab = False
        print("running locally")

    if in_colab:
        file_path = upload_file_colab()
        if file_path is None:
            file_path = "Banking_data.csv"
    else:
        file_path = input("enter the path to your CSV file (or press enter for 'Banking_data.csv'): ").strip()
        if not file_path:
            file_path = "Banking_data.csv"

    df = load_data(file_path)
    if df is None:
        return

    basic_info(df)
    analyze_age_distribution(df)
    analyze_categorical_variables(df)
    analyze_numerical_variables(df)
    analyze_target_variable(df)
    campaign_analysis(df)
    success_factors_analysis(df)
    analyze_correlations(df)

    print("analysis complete")

if __name__ == "__main__":
    main()
```
